```python
with open("saved_models/exp_18/losses_3.pkl", "rb") as f: 
    losses = pickle.load(f)
    total_gen_n2c_loss = min(losses['total_gen_n2c_loss'])
total_gen_n2c_loss
```

```python
# get losses over all epochs
epoch = 20
all_gen_n2c_loss = []
all_gen_c2n_loss =[]
all_disc_n_loss = []
all_disc_c_loss =[]
all_total_cycle_loss =[]
all_total_gen_n2c_loss = []
all_total_gen_c2n_loss = []
all_n2c_identity_loss = []
all_c2n_identity_loss = []

for i in range(epoch):
    with open("saved_models/exp_19/losses_"+str(i)+".pkl", "rb") as f: 
        losses = pickle.load(f)
        gen_n2c_loss = min(losses['gen_n2c_loss'])
        gen_c2n_loss = min(losses['gen_c2n_loss'])
        disc_n_loss = min(losses['disc_n_loss'])
        disc_c_loss = min(losses['disc_c_loss'])
        total_cycle_loss = min(losses['total_cycle_loss'])
        total_gen_n2c_loss = min(losses['total_gen_n2c_loss'])
        total_gen_c2n_loss = min(losses['total_gen_c2n_loss'])
        n2c_identity_loss = min(losses['n2c_identity_loss'])
        c2n_identity_loss = min(losses['c2n_identity_loss'])

        all_gen_n2c_loss.append(gen_n2c_loss)
        all_gen_c2n_loss.append(gen_c2n_loss)
        all_disc_n_loss.append(disc_n_loss)
        all_disc_c_loss.append(disc_c_loss)
        all_total_cycle_loss.append(total_cycle_loss)
        all_total_gen_n2c_loss.append(total_gen_n2c_loss)
        all_total_gen_c2n_loss.append(total_gen_c2n_loss)
        all_n2c_identity_loss.append(n2c_identity_loss)
        all_c2n_identity_loss.append(c2n_identity_loss)
        
```

```python
#exp_19
fig, ax = plt.subplots()
plt.plot(all_gen_n2c_loss, label='gen_n2c')
plt.plot(all_gen_c2n_loss, label='gen_c2n')
plt.plot(all_disc_n_loss, label='disc_n')
plt.plot(all_disc_c_loss, label='disc_c')

plt.plot(all_total_cycle_loss, label='total_cycle')
plt.plot(all_total_gen_n2c_loss, label='total_gen_n2c')
plt.plot(all_total_gen_c2n_loss, label='total_gen_c2n')
plt.plot(all_n2c_identity_loss, label='n2c_identity')
plt.plot(all_c2n_identity_loss, label='c2n_identity')

plt.ylabel("Loss")
plt.xlabel("Epoch")

# plt.legend(loc='upper right')

pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.8, pos.height])
ax.legend(loc='center right', bbox_to_anchor=(1.4, 0.5))
```

```python
#final
fig, ax = plt.subplots()
plt.plot(all_gen_n2c_loss, label='gen_n2c')
plt.plot(all_gen_c2n_loss, label='gen_c2n')
plt.plot(all_disc_n_loss, label='disc_n')
plt.plot(all_disc_c_loss, label='disc_c')

plt.plot(all_total_cycle_loss, label='total_cycle')
plt.plot(all_total_gen_n2c_loss, label='total_gen_n2c')
plt.plot(all_total_gen_c2n_loss, label='total_gen_c2n')
plt.plot(all_n2c_identity_loss, label='n2c_identity')
plt.plot(all_c2n_identity_loss, label='c2n_identity')

plt.ylabel("Loss")
plt.xlabel("Epoch")

# plt.legend(loc='upper right')

pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width * 0.8, pos.height])
ax.legend(loc='center right', bbox_to_anchor=(1.4, 0.5))

plt.savefig('loss_final.png')
```
